### Encode the categorical column – ‘vendor’ using label encoder

```python
# Import label encoder 
from sklearn import preprocessing 

# label_encoder object knows 
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 

# Encode labels in column 'species'. 
df['vendor']= label_encoder.fit_transform(df['vendor']) 
df
```

### b.	Identify the vendors who have manufactured less than 5 CPUs and drop those rows from the given dataset, corresponding to the identified vendors.

```python
import pandas as pd

# Assuming your DataFrame is named 'cpu_data' and the column containing vendors is named 'Vendor'

# Grouping by Vendor and counting the number of CPUs each vendor has manufactured
vendor_counts = df['vendor'].value_counts()

# Filtering vendors with less than 5 CPUs
vendors_to_drop = vendor_counts[vendor_counts < 5].index

# Filtering the DataFrame to drop the rows corresponding to these vendors
filtered_df = df[~df['vendor'].isin(vendors_to_drop)]
```

```python
filtered_df
```

### c.	Drop the column ‘model’.

```python
df.drop(['model'],axis=1,inplace=True)
```

```python
# Assuming your DataFrame is named 'cpu_data'

# Selecting 'score' as the target variable
target_variable = df['score']

# Selecting remaining columns as predictors
predictors = df.drop(['score'], axis=1)
```

### 4.	Split the data into training and testing data set in the ratio 80:20. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors, target_variable, test_size=0.2, random_state=42)
```

### 
5.	As part of model building, perform the following activities:

```python
from sklearn.linear_model import LinearRegression
# Building a Linear Regression model
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
```

### Finding the train and test scores for the model

```python
train_score = linear_regression.score(X_train, y_train)
test_score = linear_regression.score(X_test, y_test)
```
